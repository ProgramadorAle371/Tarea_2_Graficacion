Momentos HU

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os

# Función para calcular y mostrar momentos de Hu con operaciones detalladas
def calcular_momentos_hu(binaria, titulo):
    momentos = cv2.moments(binaria)
    
    # Extraer los momentos geométricos
    mu20 = momentos['mu20']
    mu02 = momentos['mu02']
    mu11 = momentos['mu11']
    mu30 = momentos['mu30']
    mu12 = momentos['mu12']
    mu21 = momentos['mu21']
    mu03 = momentos['mu03']

    print(f'\nMomentos de Hu {titulo}:')
    
    # Cálculo de φ₁
    print(f'φ₁ = μ20 + μ02 = {mu20} + {mu02}')
    phi1 = mu20 + mu02
    print(f'φ₁ = {phi1}')

    # Cálculo de φ₂
    print(f'φ₂ = (μ20 - μ02)^2 + 4 * μ11^2')
    print(f'φ₂ = ({mu20} - {mu02})^2 + 4 * ({mu11})^2')
    phi2 = (mu20 - mu02) ** 2 + 4 * mu11 ** 2
    print(f'φ₂ = {phi2}')

    # Cálculo de φ₃
    print(f'φ₃ = (μ30 - 3 * μ12)^2 + (3 * μ21 - μ03)^2')
    print(f'φ₃ = ({mu30} - 3 * {mu12})^2 + (3 * {mu21} - {mu03})^2')
    phi3 = (mu30 - 3 * mu12) ** 2 + (3 * mu21 - mu03) ** 2
    print(f'φ₃ = {phi3}')

    return phi1, phi2, phi3

# Carpeta con las imágenes
ruta = 'imagenes/'

for archivo in os.listdir(ruta):
    path = os.path.join(ruta, archivo)

    # Cargar imágenes .gif o en otros formatos
    if archivo.lower().endswith('.gif'):
        try:
            img_pil = Image.open(path).convert('L')  # Convertir GIF a escala de grises
            imagen = np.array(img_pil)
        except Exception as e:
            print(f'Error al cargar {archivo}: {e}')
            continue
    else:
        imagen = cv2.imread(path, cv2.IMREAD_GRAYSCALE)  # Leer imágenes no GIF

    # Verificar si la imagen se cargó correctamente
    if imagen is None:
        print(f'No se pudo cargar {archivo}')
        continue

    print(f'Procesando: {archivo}')

    # Binarizar la imagen
    _, binaria = cv2.threshold(imagen, 127, 255, cv2.THRESH_BINARY)

    # Mostrar la imagen binaria original
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(binaria, cmap='gray')
    plt.title(f'Original: {archivo}')
    plt.axis('off')

    # Momentos antes de rotar
    calcular_momentos_hu(binaria, 'ANTES de rotar')

    # Rotar la imagen 45 grados
    (h, w) = binaria.shape
    centro = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(centro, 45, 1.0)
    rotada = cv2.warpAffine(binaria, M, (w, h))

    # Mostrar la imagen rotada
    plt.subplot(1, 2, 2)
    plt.imshow(rotada, cmap='gray')
    plt.title('Rotada 45°')
    plt.axis('off')
    plt.show()

    # Momentos después de rotar
    calcular_momentos_hu(rotada, 'DESPUÉS de rotar')
```

Filtros

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.morphology import skeletonize
from skimage.filters import gaussian
from skimage.util import img_as_ubyte

# Celda 2: Definir funciones para los filtros
def quitar_ruido(imagen):
    return cv2.medianBlur(imagen, 5)

def suavizar_bordes(imagen):
    return cv2.GaussianBlur(imagen, (5, 5), 0)

def rellenar_huecos(imagen):
    kernel = np.ones((10, 10), np.uint8)  # Aumentamos el tamaño del kernel
    cerrada = cv2.morphologyEx(imagen, cv2.MORPH_CLOSE, kernel, iterations=2)  # Aplicamos múltiples iteraciones
    return cerrada


def encontrar_esqueletos(imagen):
    binaria = cv2.threshold(imagen, 128, 255, cv2.THRESH_BINARY)[1]
    esqueletos = skeletonize(binaria // 255)
    return img_as_ubyte(esqueletos)

# Celda 3: Función para mostrar imágenes en una fila en el notebook
def mostrar_filtros_en_fila(original, resultados, titulos):
    plt.figure(figsize=(15, 5))
    
    total_imagenes = len(resultados) + 1  # Incluye la imagen original
    
    # Mostrar imagen original
    plt.subplot(1, total_imagenes, 1)
    plt.imshow(original, cmap='gray')
    plt.title("Original")
    plt.axis('off')
    
    # Mostrar cada imagen procesada
    for i, resultado in enumerate(resultados):
        plt.subplot(1, total_imagenes, i + 2)
        plt.imshow(resultado, cmap='gray')
        plt.title(titulos[i])
        plt.axis('off')
    
    plt.show()

# Celda 4: Procesar imágenes GIF y mostrar avances con todos los filtros
def procesar_gifs_y_mostrar_todos_los_filtros(ruta_carpeta):
    archivos = [f for f in os.listdir(ruta_carpeta) if f.endswith('.gif')]
    
    if len(archivos) == 0:
        print("No se encontraron archivos GIF en la carpeta.")
        return

    for archivo in archivos:
        print(f"Procesando archivo: {archivo}")
        ruta_imagen = os.path.join(ruta_carpeta, archivo)
        gif = cv2.VideoCapture(ruta_imagen)
        ret, frame = gif.read()
        frame_count = 0
        
        while ret:
            frame_count += 1

            # Convertir a escala de grises
            frame_gris = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            
            # Aplicar los filtros
            ruido = quitar_ruido(frame_gris)
            bordes = suavizar_bordes(ruido)
            huecos = rellenar_huecos(bordes)
            esqueletos = encontrar_esqueletos(huecos)
            
            # Mostrar imágenes en una fila
            resultados = [ruido, bordes, huecos, esqueletos]
            titulos = ["Quitar Ruido", "Suavizar Bordes", "Rellenar Huecos", "Encontrar Esqueletos"]
            mostrar_filtros_en_fila(frame_gris, resultados, titulos)
            
            # Leer siguiente fotograma
            ret, frame = gif.read()
        gif.release()

# Celda 5: Ejecutar el procesamiento
ruta_carpeta = "imagenes/"  # Cambia esto por la ruta de tu carpeta
procesar_gifs_y_mostrar_todos_los_filtros(ruta_carpeta)
```
